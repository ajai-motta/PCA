```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import sklearn
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cacer_dataset=load_breast_cancer()
```

```python
cacer_dataset.keys()
```

```python
df=pd.DataFrame(cacer_dataset['data'],columns=cacer_dataset['feature_names'])
df
```

```python
from sklearn.preprocessing import StandardScaler
standard=StandardScaler()
```

```python
standard.fit(df)
```

```python
scaled_data=standard.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA 
pca=PCA(n_components=2) #n_componets is saying how much we want to reduce dimesionality to
```

```python
pca_data=pca.fit_transform(scaled_data)
pca_data
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```
