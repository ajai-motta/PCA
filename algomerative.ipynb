```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import sklearn
%matplotlib inline
```

```python
from sklearn import datasets
iris=datasets.load_iris()
iris
```

```python
iris_df=pd.DataFrame(iris.data)
iris_df
```

```python
iris_df.columns=iris.feature_names
iris_df
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
iris_scaled=scalar.fit_transform(iris_df)
```

```python

iris_scaled
```

```python
#dimensionality reducyion using pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
reduced_iris=pca.fit_transform(iris_scaled)
```

```python
plt.scatter(reduced_iris[:,0],reduced_iris[:,1],c=iris.target)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,7))
plt.title("dendogram")
sc.dendrogram(sc.linkage(reduced_iris,method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
```

```python
clustering.fit(reduced_iris)
```

```python
plt.scatter(reduced_iris[:,0],reduced_iris[:,1],c=clustering.labels_)
plt.show()
```
